## Elementary School Processing

```python
out_df = pd.DataFrame()
def select_valid_columns(in_df,out_df):
    in_df = in_df[['locname','citystzi','SC17_Eng4EXp','SC17_Eng3REp','SC17_Eng2CLp','SC17_Eng1NSp',
              'SC17_Mat4EXp','SC17_Mat3REp','SC17_Mat2CLp','SC17_Mat1NSp',
               'P17_Sci7EXC','P17_Sci5EXE','P17_Sci3Met','P17_Sci1NM'
                ]]
    print(in_df.head())
    return 

elem_file = "RC_2017_Elementary.csv"
in_df = pd.read_csv(elem_file,delimiter=",")

# elem_df = elem_df[['locname','citystzi','SC17_Eng4EXp','SC17_Eng3REp','SC17_Eng2CLp','SC17_Eng1NSp',
#               'SC17_Mat4EXp','SC17_Mat3REp','SC17_Mat2CLp','SC17_Mat1NSp',
#                'P17_Sci7EXC','P17_Sci5EXE','P17_Sci3Met','P17_Sci1NM'
#                 ]]

select_valid_columns(in_df,out_df)
out_df
```

```python
elem_df['citystzi'] = elem_df['citystzi'].astype(str).str[-5:]
elem_df.head()
```

```python
elem_df["English pass"] = elem_df["SC17_Eng4EXp"] + elem_df["SC17_Eng3REp"] + elem_df["SC17_Eng2CLp"]

elem_df["Math pass"] = elem_df["SC17_Mat4EXp"] + elem_df["SC17_Mat3REp"] + elem_df["SC17_Mat2CLp"]

elem_df["Science pass"] = elem_df["P17_Sci7EXC"] + elem_df["P17_Sci5EXE"] + elem_df["P17_Sci3Met"]


elem_df["Overall pass"] = round((elem_df["English pass"] + elem_df["Math pass"] + elem_df["Science pass"]) / 3,2)

elem_final = elem_df[['locname','citystzi','English pass','SC17_Eng1NSp','Math pass',
                      'SC17_Mat1NSp',"Science pass","P17_Sci1NM","Overall pass"]]

elem_final = elem_final.rename(columns = {'citystzi':'zip_code','SC17_Eng1NSp':'English fail',
                                          'SC17_Mat1NSp':'Math fail','locname':'School','P17_Sci1NM':'Science fail'})

elem_final["rating"] = round(elem_final["Overall pass"] / 10,0)

elem_final.head()
```

```python
elem_rating_for_available_zip = pd.merge(elem_final,zip_df,on="zip_code")
elem_rating_for_available_zip = elem_rating_for_available_zip[["city","zip_code","School","English pass","Math pass",
                                                               "Science pass","Overall pass","rating"]]
elem_rating_for_available_zip.head()
```
