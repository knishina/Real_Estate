```python
#import libraries
import pandas as pd
```

```python
zip_file ="zipcodes.csv"
zip_df = pd.read_csv(zip_file,delimiter=",")
zip_df['zip'] = zip_df['zip'].astype(str)
zip_df.head()
```

```python
def select_the_required_columns(input_df):
    organized_df = input_df[['locname','TYPE','citystzi','SC17_Eng4EXp','SC17_Eng3REp','SC17_Eng2CLp','SC17_Eng1NSp',
              'SC17_Mat4EXp','SC17_Mat3REp','SC17_Mat2CLp','SC17_Mat1NSp',
               'P17_Sci7EXC','P17_Sci5EXE','P17_Sci3Met','P17_Sci1NM'
                ]]
    
    organized_df['citystzi'] = organized_df['citystzi'].astype(str).str[-5:]
    return organized_df


######################################################################################################################

def calculate_values(input_df):
    
    input_df["English pass"] = input_df["SC17_Eng4EXp"] + input_df["SC17_Eng3REp"] + input_df["SC17_Eng2CLp"]
    input_df["Math pass"] = input_df["SC17_Mat4EXp"] + input_df["SC17_Mat3REp"] + input_df["SC17_Mat2CLp"]
    input_df["Science pass"] = input_df["P17_Sci7EXC"] + input_df["P17_Sci5EXE"] + input_df["P17_Sci3Met"]
    input_df["Overall pass"] = round((input_df["English pass"] + input_df["Math pass"] + input_df["Science pass"]) / 3,2)
    
    output_df = input_df[['locname','TYPE','citystzi','English pass','SC17_Eng1NSp','Math pass',
                      'SC17_Mat1NSp',"Science pass","P17_Sci1NM","Overall pass"]]
    
    output_df = output_df.rename(columns = {'citystzi':'zip','SC17_Eng1NSp':'English fail','TYPE':'Type',
                                          'SC17_Mat1NSp':'Math fail','locname':'School','P17_Sci1NM':'Science fail'})
    
    output_df["rating"] = round(output_df["Overall pass"] / 10,0)
    return output_df

######################################################################################################################

def calculate_rating_for_available_zip(input_df):
    output_rating_available_zip = pd.merge(zip_df,input_df,on="zip",how="left")
#     output_rating_available_zip = output_rating_available_zip[["primary_city","zip","School","Type","English pass","Math pass",
#                                                                "Science pass","Overall pass","rating"]]
    output_rating_available_zip = output_rating_available_zip[["primary_city","zip","School","Type","rating"]]
    return output_rating_available_zip
```

## Elementary School Processing

```python
elementary_file = "RC_2017_Elementary.csv"
input_df = pd.read_csv(elementary_file,delimiter=",")

elementary_df = select_the_required_columns(input_df)
elementary_final = calculate_values(elementary_df)
elementary_rating_for_available_zip = calculate_rating_for_available_zip(elementary_final)
elementary_rating_for_available_zip.head()
```

## Middle School Processing

```python
middle_file = "RC_2017_Middle.csv"
input_df = pd.read_csv(middle_file,delimiter=",")

middle_df = select_the_required_columns(input_df)
middle_final = calculate_values(middle_df)
middle_rating_for_available_zip = calculate_rating_for_available_zip(middle_final)
middle_rating_for_available_zip.head()
```
